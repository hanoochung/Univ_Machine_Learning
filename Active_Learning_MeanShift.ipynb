```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=10, n_features=2, centers=2, random_state=3)

plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
display(np.round(X, 1))
```
